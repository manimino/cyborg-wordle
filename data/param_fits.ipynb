```python
def summary(df):
    for col in ['exponent', 'pool_size', 'ratio']:
        print('')
        print(col)
        for v in sorted(set(df[col].to_list())):
            dfv = df[df[col] == v]
            print(v, sum(dfv['avg']))
```
